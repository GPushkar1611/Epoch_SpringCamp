```python
import numpy as np
import matplotlib.pyplot as plt
```

## Task 1

```python
def generate_grayscale_image(height, width):
    return np.random.randint(0, 256, (height, width))
```

```python
def four_way_kaleidoscope(image):
    h, w = image.shape
    # Divide the image into quadrants
    top_left = image[:h//2, :w//2]  # Original top-left quadrant

    # Generate other quadrants using mirroring and flipping
    top_right = np.fliplr(top_left)  # Mirror left → right
    bottom_left = np.flipud(top_left)  # Flip top → bottom
    bottom_right = np.flipud(top_right)  # Flip top → bottom

    # Combine the quadrants to form the final image
    top_half = np.hstack((top_left, top_right))
    bottom_half = np.hstack((bottom_left, bottom_right))
    kaleidoscope_image = np.vstack((top_half, bottom_half))

    return kaleidoscope_image
```

```python
original_image = generate_grayscale_image(50,50)

# Apply the four way kaleidoscope transformation
kaleidoscope_image = four_way_kaleidoscope(original_image)

# Display the original and transformed images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(original_image, cmap='gray')
axes[0].set_title("Original Image")

axes[1].imshow(kaleidoscope_image, cmap='gray')
axes[1].set_title("Four Way Kaleidoscope Image")

plt.show()
```

```python
def generate_rgb_image(height, width):
    return np.random.randint(0, 256, (height, width, 3))
```

```python
def four_way_kaleidoscope_rgb(image):
    h, w, c = image.shape  # Get image dimensions (height, width, channels)
    transformed_channels = []

    for i in range(c):  # Loop over R, G, B channels
        channel = image[:, :, i]  # Extract the current channel
        transformed_channel = four_way_kaleidoscope(channel)  # Apply kaleidoscope transformation
        transformed_channels.append(transformed_channel)  # Store the transformed channel

    # Stack the transformed channels back to form the final RGB image
    kaleidoscope_image = np.dstack(transformed_channels)

    return kaleidoscope_image
```

```python
original_image = generate_rgb_image(25,25)

# Apply the four way kaleidoscope transformation
kaleidoscope_image = four_way_kaleidoscope_rgb(original_image)

# Display the original and transformed images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(original_image)
axes[0].set_title("Original RGB Image")

axes[1].imshow(kaleidoscope_image)
axes[1].set_title("Four Way Kaleidoscope RGB Image")

plt.show()
```
